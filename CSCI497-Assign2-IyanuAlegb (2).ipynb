```python
# Remove Outliers
import numpy as np

def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return df[(df[column] >= lower) & (df[column] <= upper)]

# Apply for GDP
data = remove_outliers(data, 'GDP')

# Apply for Population
data = remove_outliers(data, 'Population')
```

```python
data.plot.box()
```

```python
# Remove outliers column by column to get a better graph

def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)   # 25th percentile
    Q3 = df[column].quantile(0.75)   # 75th percentile
    IQR = Q3 - Q1                    # Interquartile Range
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    # Keep only values within the bounds
    return df[(df[column] >= lower) & (df[column] <= upper)]

cleaned = data.copy()

# Remove outliers in GDP
cleaned = remove_outliers(cleaned, 'GDP')

# Remove outliers in Population
cleaned = remove_outliers(cleaned, 'Population')

# Remove outliers in Alcohol
cleaned = remove_outliers(cleaned, 'Alcohol')

# Remove outliers in Alcohol
cleaned = remove_outliers(cleaned, 'Schooling')

```

## Explanation Correlation of Features

According to the correlation heat map, multiple variables in the dataset have robust correlations with each other. For instance, infant deaths and under-five deaths are strongly positively correlated because they both measure child mortality. Similarly, thinness 1â€“19 years and thinness 5â€“9 years are almost the same: they represent malnutrition at different ages. Features related to vaccination, for example, Polio, Diphtheria, and Hepatitis B, are also highly correlated because those countries with strong immunization programs generally perform well across all vaccines. In other cases, however, some variables are negatively associated with life expectancy. For example, Adult Mortality and the prevalence of HIV/AIDS are negatively correlated with life expectancy, indicating that as these values increase, life expectancy decreases. In contrast, for example, Schooling, GDP, and health expenditure all correlated positively with life expectancy (higher education, stronger economies, and increasing expenditures on healthcare all contribute to longer life expectancy). These correlations indicate that not all features are unique. Features with high correlations have the potential to lead to redundancy and multicollinearity in regression, and several of them were dropped (for example, under-five deaths and thinness 5â€“9 years).

```python
# Skewness / Normality Check with Shapiro-Wilk
from scipy.stats import shapiro

# Loop through numeric columns in your cleaned dataset
numeric_cols = final_data.select_dtypes(include=['float64','int64']).columns

for col in numeric_cols:
    stat, p = shapiro(final_data[col].dropna())   # drop NaN to avoid errors
    print(f"{col}: Statistics={stat:.3f}, p={p:.3f}")
    
    if p > 0.05:
        print("  â†’ Data looks Gaussian (normal distribution)")
    else:
        print("  â†’ Data does not look Gaussian (skewed)")
```
